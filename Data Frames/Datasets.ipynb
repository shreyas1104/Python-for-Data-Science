```python
# Filtering Data frames, filtering is about rows
```

```python
stats.head()
```

```python
stats.InternetUsers < 2               #rows in which this condition is met
```

```python
filter = stats.InternetUsers < 2
```

```python
filter
```

```python
stats[filter]          #shows only rows where filter is true, similar in R
```

```python
filter2 = stats.BirthRate > 40
```

```python
stats[filter2]
```

```python
stats[stats.BirthRate > 40]          #can put th filter inline instead of assigning to a variable
```

```python
# Combining filters
stats[stats.BirthRate > 40 and stats.InternetUsers < 2]
```

```python
stats.BirthRate > 40 and stats.InternetUsers < 2         #and will not work with such objects
```

```python
stats[(stats.BirthRate > 40) & (stats.InternetUsers < 2)]            #brackets are important, will error without it
```

```python
stats[filter & filter2]
```

```python
stats[stats.IncomeGroup == 'High income']
```

```python
stats[(stats.IncomeGroup == 'High income') & (stats.InternetUsers < 40)]
```

```python
# how to get unique categories
stats.IncomeGroup.unique()
```

```python
#Find out everything about Malta
stats[stats.CountryName == 'Malta']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams['figure.figsize'] = 8,4
import warnings
warnings.filterwarnings('ignore')
```

```python
stats.head()
```

```python
#Distribution
vis1 = sns.distplot(stats['InternetUsers'])
```

```python
vis1 = sns.distplot(stats['InternetUsers'], bins=30)
```

```python
#Boxplots
vis2 = sns.boxplot(data=stats, x='IncomeGroup', y='BirthRate')
```

```python
#Seaborn Gallery
```

---

```python
vis5 = sns.lmplot(x='InternetUsers', y='BirthRate', data=stats )   # can specify parameters in any order by mentioning names with values
```

```python
vis3 = sns.lmplot(x='InternetUsers', y='BirthRate', data=stats, fit_reg=False )      # fit_reg false removes the line
```

```python
vis3 = sns.lmplot(x='InternetUsers', y='BirthRate', data=stats, \
                  fit_reg=False, hue='IncomeGroup', size=8 )              # hue adds color based on IncomeGroup, size is used to increase/decrease size
```

```python
# Keyword Arguments in Python
```

```python
# changing marker color
vis3 = sns.lmplot(x='InternetUsers', y='BirthRate', data=stats, \
                  fit_reg=False, hue='IncomeGroup', size=8, \
                 scatter_kws={'s':100})            #scatter_kws is a dictionary for additional arguments using pyplot
```
