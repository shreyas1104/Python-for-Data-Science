```python
# using loops to get each genre
genre_list = []
myLabels = []
for gen in movies.Genre.cat.categories:
    myLabels.append(gen)
    genre_list.append(movies[movies.Genre == gen].BudgetMillions)       #creating a list of budgets for each genre

plt.hist(genre_list, bins=30, stacked=True, rwidth=1, label=myLabels)
plt.legend()
plt.show()
```

```python
g = sns.FacetGrid(data=movies, row='Genre', col='Year', hue='Genre')
g = g.map(plt.scatter, 'CriticRating', 'AudienceRating')
```

```python
#can populate with any type of chart, say histograms
g = sns.FacetGrid(data=movies, row='Genre', col='Year', hue='Genre')
g = g.map(plt.hist, 'BudgetMillions')
```

```python
# adding keyword arguments
g = sns.FacetGrid(data=movies, row='Genre', col='Year', hue='Genre')
kws = dict(s=50, linewidth=0.5, edgecolor='black')        #using keyword arguments
g = g.map(plt.scatter, 'CriticRating', 'AudienceRating', **kws)
```

```python
# Controlling axes and adding diagonals
g = sns.FacetGrid(data=movies, row='Genre', col='Year', hue='Genre')
kws = dict(s=50, linewidth=0.5, edgecolor='black')
g = g.map(plt.scatter, 'CriticRating', 'AudienceRating', **kws)
g.set(xlim=(0,100), ylim=(0,100))              #setting axes to start from 0 and go to 100

#adding diagonals, to understand whether critics or audiences gave better ratings
for ax in g.axes.flat:
    ax.plot((0,100), (0,100), c='gray', ls='--')

g.add_legend()
```

```python
# Building Dashboards
```

```python
sns.set_style('darkgrid')
f, axes = plt.subplots(2, 3, figsize=(15,15))
k1 = sns.kdeplot(movies.BudgetMillions, movies.AudienceRating, ax=axes[0][0])
k1 = sns.kdeplot(movies.BudgetMillions, movies.CriticRating, ax=axes[0][1])
w2 = sns.violinplot(data=movies[movies.Genre == 'Drama'], x='Year', y='CriticRating', ax=axes[1,0])
b2 = sns.boxplot(data=movies[movies.Genre == 'Drama'], x='Year', y='CriticRating', ax=axes[1,1])
axes[0,2].hist(movies.CriticRating, bins=15)            #adding a non seaborn plot to subplot, these do not have ax parameter,
            #this will not work with sns plots since axes is an object of pyplot, and sns plots have an added parameter of ax
axes[1,2].hist(movies.AudienceRating, bins=15)
plt.show()

#axes[x][y] and axes[x,y] both work
```
