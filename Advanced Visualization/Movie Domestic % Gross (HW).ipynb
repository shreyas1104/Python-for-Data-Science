```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
movies = pd.read_csv('Movie-Homework-Dataset.csv')
```

```python
movies.head()
```

```python
movies.columns
```

```python
movies.columns = ['DayofWeek', 'Director', 'Genre', 'MovieTitle', 'ReleaseDate', \
               'Studio', 'AdjustedGross', 'Budget)', 'Gross', \
               'IMDb', 'MovieLensRating', 'Overseas', 'OverseasPercent', \
               'Profit', 'Profit%', 'Runtime', 'US','GrossUsPercent']
```

```python
movies.head()
```

```python
# extract only the required genres
genre_filter = (movies.Genre == 'action') | (movies.Genre == 'comedy') | (movies.Genre == 'adventure') | \
(movies.Genre == 'animation') | (movies.Genre == 'drama')
```

```python
# extract only the required studios
studio_filter = (movies.Studio == 'Buena Vista Studios') | (movies.Studio == 'Sony') | \
(movies.Studio == 'Universal') | (movies.Studio == 'WB') | (movies.Studio == 'Paramount Pictures') | \
(movies.Studio == 'Fox')
```

```python
len(movies[genre_filter & studio_filter])
```

```python
sns.set_style('darkgrid')
f, ax = plt.subplots()
f.set_size_inches(11.7, 8.27)

#Boxplot
bp = sns.boxplot(data=movies[genre_filter & studio_filter], x='Genre', y='GrossUsPercent', showfliers=False, color='lightgray')

#Stripplot
jt = sns.stripplot(data=movies[genre_filter & studio_filter], x='Genre', y='GrossUsPercent', hue='Studio')

plt.title('Domestic Gross % by Genre', fontsize=30)
plt.ylabel('Gross US %', fontsize=20)
plt.xlabel('Genre', fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend(fontsize=15, frameon=True, fancybox=True, shadow=True, loc='upper left', bbox_to_anchor=(1,1))
plt.show()
```
